```python
# Function to remove price per sqft outliers per location using mean and 1 standard deviation
def outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)    # mean
        st = np.std(subdf.price_per_sqft)    # standard deviation
        reduced_df = subdf[(subdf.price_per_sqft > (m - st)) & (subdf.price_per_sqft <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out
```

```python
df = outliers(df)
```

```python
df
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        BHK_stats = {}
        for BHK, BHK_df in location_df.groupby('BHK'):
            BHK_stats[BHK] = {
                'mean': np.mean(BHK_df.price_per_sqft),
                'std': np.std(BHK_df.price_per_sqft),
                'count': BHK_df.shape[0]
            }
        for BHK, BHK_df in location_df.groupby('BHK'):
            stats = BHK_stats.get(BHK-1)
            if stats and stats['count']>5:
                exclude_indices = np.append(exclude_indices, BHK_df[BHK_df.price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')
```

```python
df = remove_bhk_outliers(df)
```

# MODEL BUILDING

```python
X = df.drop("price",axis=1)
y = df["price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=12)
```

```python
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import OneHotEncoder,StandardScaler
```

```python
col_transform = make_column_transformer((OneHotEncoder(sparse_output=False),["location"]),remainder="passthrough")
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

# Applying Linear Regression

```python
pipe = make_pipeline(col_transform,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test,y_pred)
```

```python
r2
```
